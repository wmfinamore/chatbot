```python
def preprocessamento(texto):
    # substitui URLs por espaço em branco
    texto = re.sub(r'https?://[A-Za-z0-9./]+', '', texto)
    
    # espaços em branco
    texto = re.sub(r' +', ' ', texto)
    
    documento = pln(texto)
    lista = []
    for token in documento:
        lista.append(token.lemma_)
        
    lista = [palavra for palavra in lista if palavra not in stop_words and palavra not in string.punctuation and palavra not in ('\n','\n ', '\t', '\t\t', ' \n\t\t', '\n\n')]
    
    lista = ' '.join([str(elemento) for elemento in lista if not elemento.isdigit()])
    
    return lista
```

```python
texto_teste = 'https://www.sorocaba.sp.gov.br '+lista_sentencas[0]
texto_teste
```

```python
resultado = preprocessamento(texto_teste)
resultado
```

```python
lista_sentencas_preprocessada = []

for i in range(len(lista_sentencas)):
    lista_sentencas_preprocessada.append(preprocessamento(lista_sentencas[i]))
```

```python
textos_boas_vindas_entradas = ('hey','olá','opa','oi','eae','bom dia')
textos_boas_vindas_respostas = ('hey','olá','opa','oi','bem vindo','como você está?')
```

```python
def responder_saudacao(texto):
    for palavra in texto.split():
        if palavra.lower() in textos_boas_vindas_entradas:
            return random.choice(textos_boas_vindas_respostas)
```

```python
responder_saudacao('olá tudo bem?')
```
